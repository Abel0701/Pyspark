```
from pyspark.sql.functions import *
```

##Leemos el archivo

```
df = spark.read.format("csv")\
    .option("header", "true")\
    .option("inferschema", "true")\
    .load("/FileStore/shared_uploads/arraz7191@gmail.com/ds_salaries.csv")
```

##Mostramos la tabla

```
df.show()
```

##Cantidad de registros

```
df.count()
```

##Salario promedio

```
salario_promedio = df.select(avg("salary")).collect()[0][0]
print("El salario promedio es: ", salario_promedio)
```

##Filtramos datos a partir de un salario

```
empleados_con_salario_mayor = df.filter(df["salary"] > 100000).orderBy(col("salary").desc())
empleados_con_salario_mayor.show()
```

##Seleccionamos columnas específicas, filtramos por moneda y ordenamos el salario de mayor a menor

```
df.select(
    col("job_title").alias("Puesto"), 
    col("salary").alias("Salario"), 
    col("salary_currency").alias("Moneda"),
    col("employee_residence").alias("Lugar de residencia")                                               
).filter(df["salary_currency"] == "USD").orderBy(col("salary").desc()).show()
```
